```python
import pandas as pd
from docxtpl import DocxTemplate
from docx import Document

# Load the Word template
template_path = "Resultcard.docx"
template_doc = DocxTemplate(template_path)

# Load the student data from the CSV file
csv_path = "R.csv"
df = pd.read_csv(csv_path)

# Create a new Word document to hold all mark sheets
output_doc = Document()

# Generate a mark sheet for each student
for index, row in df.iterrows():
    # Define the context with the student's data
    context = {
        'GR_No': row['GR_No'],
        'NAME': row['NAME'],
        'F_NAME': row['F_NAME'],
        'Class': row['Class'],
        'English': row['English'],
        'Math': row['Math'],
        'Scien': row['Scien'],
        'IsI': row['IsI'],
        'Urdu': row['Urdu'],
        'S_S_T': row['S_S_T'],
        'Draw': row['Draw'],
        'Spe': row['Spe'],
        'had': row['had'],
        'Total': row['Total'],
        'Per': row['Per'],
        'Grade': row['Grade']
    }

    # Render the template with the current student's data
    template_doc.render(context)

    # Save the rendered template as a temporary document
    temp_file = "temp.docx"
    template_doc.save(temp_file)

    # Load the rendered template and append it to the final document
    temp_doc = Document(temp_file)
    for element in temp_doc.element.body:
        output_doc.element.body.append(element)

    # Add a page break between students (except for the last one)
    if index < len(df) - 1:
        output_doc.add_page_break()

# Save the combined document
output_filename = "mark_sheets.docx"
output_doc.save(output_filename)

print(f"Generated: {output_filename}")
```
